## 2. 계절별로 5개 시즌 데이터 합치기

```python
type(spring_w_15)
```

```python
sprW15 = pd.DataFrame(spring_w_15)
sprW16 = pd.DataFrame(spring_w_16)
sprW17 = pd.DataFrame(spring_w_17)
sprW18 = pd.DataFrame(spring_w_18)
sprW19 = pd.DataFrame(spring_w_19)
```

```python
sumW15 = pd.DataFrame(summer_w_15)
sumW16 = pd.DataFrame(summer_w_16)
sumW17 = pd.DataFrame(summer_w_17)
sumW18 = pd.DataFrame(summer_w_18)
sumW19 = pd.DataFrame(summer_w_19)

fallW15 = pd.DataFrame(fall_w_15)
fallW16 = pd.DataFrame(fall_w_16)
fallW17 = pd.DataFrame(fall_w_17)
fallW18 = pd.DataFrame(fall_w_18)
fallW19 = pd.DataFrame(fall_w_19)
```

```python
sprW = pd.concat([sprW15,sprW16,sprW17,sprW18,sprW19])
sumW = pd.concat([sumW15,sumW16,sumW17,sumW18,sumW19])
fallW = pd.concat([fallW15,fallW16,fallW17,fallW18,fallW19])
```

```python
grouped_spr = sprW['Win'].groupby(sprW.index).sum()
grouped_spr = pd.DataFrame(grouped_spr)
grouped_spr
```

```python
grouped_sum = sumW['Win'].groupby(sumW.index).sum()
grouped_sum = pd.DataFrame(grouped_sum)
grouped_fall = fallW['Win'].groupby(fallW.index).sum()
grouped_fall = pd.DataFrame(grouped_fall)
grouped_fall
```

```python
grouped_spr.sort_values(by='Win',ascending=False)
```

```python
grouped_sum.sort_values(by='Win',ascending=False)
```

```python
grouped_fall.sort_values(by='Win',ascending=False)
```
