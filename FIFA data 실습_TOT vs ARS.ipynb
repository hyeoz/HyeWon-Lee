```python
#주전선수 비교를 위해 포지션별 주전 필터링

gk_list = ['GK']
cb_list = ['CB', 'LCB', 'RCB', 'RB', 'LB', 'RWB','LWB']
mf_list = ['RCM', 'LCM', 'RDM', 'CDM','LDM', 'CM','RM','LM','CAM','LAM', 'RAM']
st_list = ['ST', 'LW', 'RW', 'CF','LF','RF','LS','RS']
```

선수들 간 능력을 비교할 수 있는 항목(나만의 계산법) 공식 만들어보기

ability = Overall + Potential / Age

```python
tot['Ability'] = tot['Overall'] + (tot['Potential'] // tot['Age'])
ars['Ability'] = ars['Overall'] + (ars['Potential'] // ars['Age'])
tot[['Name','Age','Overall','Potential','Ability','Position']]
```

```python
tot = tot.sort_values('Ability', ascending = False)
ars = ars.sort_values('Ability', ascending = False)
```

```python

st_count = 2
mf_count = 4
cb_count = 4
gk_count = 1

tot_id = []

for index in tot.index:
    if tot['Position'][index] in gk_list:
        if gk_count != 0:
            tot_id.append(tot['ID'][index])
            gk_count -= 1
    elif tot['Position'][index] in mf_list:
        if mf_count != 0:
            tot['Position'][index] = 'MF'
            tot_id.append(tot['ID'][index])
            mf_count -= 1
    elif tot['Position'][index] in cb_list:
        if cb_count != 0:
            tot['Position'][index] = 'CB'
            tot_id.append(tot['ID'][index])
            cb_count -= 1
    elif tot['Position'][index] in st_list:
        if st_count != 0:
            tot['Position'][index] = 'ST'
            tot_id.append(tot['ID'][index])
            st_count -= 1

            
```

```python
tot_s = tot[tot['ID'].isin(tot_id)][['Name','Age','Ability','Overall','Potential','Position','Club','Value','Wage']]
```

```python
#예상 이적료와 주급 단위 고치고 숫자형으로 바꾸기
tot_s['Value'] = tot_s['Value'].str.replace('M', '000000')
tot_s['Value'] = tot_s['Value'].str.replace('K', '000')
tot_s['Value'] = tot_s['Value'].str.slice(1,)
tot_s['Value'] = tot_s['Value'].str.replace('.','')
tot_s['Value'] = tot_s['Value'].astype(int)
```

```python
tot_s['Wage'] = tot_s['Wage'].str.replace('M', '000000')
tot_s['Wage'] = tot_s['Wage'].str.replace('K', '000')
tot_s['Wage'] = tot_s['Wage'].str.slice(1,)
tot_s['Wage'] = tot_s['Wage'].str.replace('.','')
tot_s['Wage'] = tot_s['Wage'].astype(int)
```

```python

st_count = 2
mf_count = 4
cb_count = 4
gk_count = 1

ars_id = []

for index in ars.index:
    if ars['Position'][index] in gk_list:
        if gk_count != 0:
            ars_id.append(ars['ID'][index])
            gk_count -= 1
    elif ars['Position'][index] in mf_list:
        if mf_count != 0:
            ars['Position'][index] = 'MF'
            ars_id.append(ars['ID'][index])
            mf_count -= 1
    elif ars['Position'][index] in cb_list:
        if cb_count != 0:
            ars['Position'][index] = 'CB'
            ars_id.append(ars['ID'][index])
            cb_count -= 1
    elif ars['Position'][index] in st_list:
        if st_count != 0:
            ars['Position'][index] = 'ST'
            ars_id.append(ars['ID'][index])
            st_count -= 1

```

```python
ars_s = ars[ars['ID'].isin(ars_id)][['Name','Age','Ability','Overall','Potential','Position','Club','Value','Wage']]
```

```python
ars_s['Value'] = ars_s['Value'].str.replace('M', '000000')
ars_s['Value'] = ars_s['Value'].str.replace('K', '000')
ars_s['Value'] = ars_s['Value'].str.slice(1,)
ars_s['Value'] = ars_s['Value'].str.replace('.','')
ars_s['Value'] = ars_s['Value'].astype(int)
```

```python
ars_s['Wage'] = ars_s['Wage'].str.replace('M', '000000')
ars_s['Wage'] = ars_s['Wage'].str.replace('K', '000')
ars_s['Wage'] = ars_s['Wage'].str.slice(1,)
ars_s['Wage'] = ars_s['Wage'].str.replace('.','')
ars_s['Wage'] = ars_s['Wage'].astype(int)
```

```python
import seaborn as sns
vs_s = pd.concat([tot_s, ars_s])
```

```python
sns.boxplot(data = vs_s, x = 'Position',  y = 'Ability', hue = 'Club')
```

## 아스날이 토트넘보다 좋은 성적을 거두기 위해서 보완해야 할 포지션 중 가장 시급한 포지션은 센터백.

```python
sns.boxplot(data = vs_s, x = 'Position', y = 'Value', hue = 'Club')
```

이적료와 주급을 확인했을 때 교체가 시급한 아스날의 센터백 예상 이적료가 높고, 주급이 비교적 낮은 것을 알 수 있다.

또 주급은 토트넘이 비교적 높은데, 선수를 데려온 후 내부 계약을 통해 주급을 높여가며 선수에게 동기부여 하는 것이 선수의 성적에 영향을 줄 수 있다는 것을 예상할 수 있다.
