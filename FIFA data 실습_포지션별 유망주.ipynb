# 1. 최고 유망주 3명



포지션별 3명씩 뽑아보기
1. gk_list = ['GK']
2. cb_list = ['CB', 'LCB', 'RCB', 'RB', 'LB']
3. mf_list = ['RCM', 'LCM', 'RDM', 'CDM', 'CM','RM', 'CAM']
4. st_list = ['ST', 'LW', 'RW']

유망주의 기준 - 나이 : 16~20세

선수 가치 - 오버롤 * 1/3 + 포텐셜 * 2/3 

: 포텐셜에 더 중점을 두고 계산.

```python
# 20세 이하 선수들만 뽑기.
hope = data[data['Age'] <= 20]
#hope.info() 
hope.head()
#총 3327명
```

```python
#오버롤과 포텐셜을 이용한 새로운 컬럼 만들기
pd.set_option('display.max_columns', 80)
hope['OP'] = (hope['Overall'] * 1/3) + (hope['Potential'] * 2/3)
hope.head()
```

```python
#OP로 정렬. 이상치/결측치 보정
hope.sort_values('OP', ascending = False).head(20)
pd.set_option('display.max_columns', 80)
hope
```

```python
# 나이가 19살임에도 Overall이 150인것을 확인하였으면, 어떻게 처리할지 생각해 봅시다.
pd.set_option('display.max_columns', 80)
hope[hope['Overall']>100]
```

```python
# 같은 포지션의 선수를 살펴보니, CB포지션의 점수가 같으면, Overall이 같은 것을 볼 수 있습니다.
hope[hope['Position'] == 'CB'][['Position', 'Overall','CB']].sort_values('Overall', ascending = False)
hope['Overall'][11422] = 75
```

```python
hope.head()
hope.sort_values('OP', ascending = False)[['Name', 'Position']]
```

```python
#포지션별로 묶어주기
gk_list = ['GK']
cb_list = ['CB', 'LCB', 'RCB', 'RB', 'LB']
mf_list = ['RCM', 'LCM', 'RDM', 'CDM', 'CM','RM', 'CAM']
st_list = ['ST', 'LW', 'RW']
```

```python
#OP 기준으로 내림차순 정렬(각 포지션별로 OP기준 3명씩 뽑아야함)
hope = hope.sort_values(by = ['Position','OP'],ascending = False)

#각 포지션별 3명씩으로 제한
st_count = 3
mf_count = 3
cb_count = 3
gk_count = 3

#뽑힌 유망주 ID기준으로 리스트 만들기
hope_gk_id = []
hope_cb_id = []
hope_mf_id = []
hope_st_id = []

#반복문으로 유망주 뽑기
for index in hope.index :
    if hope['Position'][index] in gk_list :
        if gk_count != 0:
            hope_gk_id.append(hope['ID'][index])
            gk_count -= 1
    elif hope['Position'][index] in cb_list :
        if cb_count != 0:
            hope['Position'][index] = 'CB'
            hope_cb_id.append(hope['ID'][index])
            cb_count -= 1
    elif hope['Position'][index] in mf_list :
        if mf_count != 0:
            hope['Position'][index] = 'MF'
            hope_mf_id.append(hope['ID'][index])
            mf_count -= 1
    elif hope['Position'][index] in st_list:
        if st_count != 0:
            hope['Position'][index] = 'ST'
            hope_st_id.append(hope['ID'][index])
            st_count -=1


```

```python
#각 3명 뽑혔는지 확인
hope_mf_id
```

```python
#isin 함수 이용하여 뽑힌 선수 데이터 불러오기
hope_gk = hope[hope['ID'].isin(hope_gk_id)]
hope_cb = hope[hope['ID'].isin(hope_cb_id)]
hope_mf = hope[hope['ID'].isin(hope_mf_id)]
hope_st = hope[hope['ID'].isin(hope_st_id)]
```
