### 1. 검색 후 나오는 첫 페이지 크롤링으로 코드 짜보기

1) 링크 불러오기

```python
import requests
from bs4 import BeautifulSoup
```

```python
import urllib.request 
import urllib.parse 

#검색 키워드가 한글, 링크에 한글이 깨져 나옴 -> 변수로 만들어 유니코드로 치환
value = "로고" 

#parse모듈을 사용해 한글("한글")변수를 유니코드로 치환 
encode = urllib.parse.quote_plus(value) 
```

```python
#기본 갤러리 주소
url = "https://gall.dcinside.com/board/lists?id=skwyverns"

#검색어 등 추가 파라미터, 개발자 도구에서 헤더 확인(숨겨진 코드 확인)
params = {"s_type":"search_subject_memo","s_keyword":encode}
headers = {"User-Agent":"-"}
```

```python
#성공적으로 불러오는지 확인(Response 200)
resp = requests.get(url,params=params,headers=headers)
resp
```

```python
resp.url
```

2) 불러온 링크로 파싱 시작

```python

parse = BeautifulSoup(resp.content, "html.parser")
parse
```

```python
content = parse.find("tbody")
content
```

```python
list_tag = content.find_all("tr")
list_tag
```

3) 데이터에서 확인하고싶은 제목과 조회수, 추천수만 텍스트로 추출하기

```python
#전체에서 텍스트만 추출하면 에러, 반복문으로 한개씩 뽑아서 택스트 추출.
rogo = []

for i in list_tag:
    if i.find("em",{"class":"icon_img icon_notice"}):
        pass
    elif i.find("em",{"class":"icon_img icon_survey"}):
        pass
    else:
        print("-" * 10)

        title_tag = i.find("a")
        title = title_tag.text
        print("제목:", title)

        count_tag = i.find("td",{"class":"gall_count"})
        count = count_tag.text
        print("조회수:", count)

        recommend_tag = i.find("td",{"class":"gall_recommend"})
        recom = recommend_tag.text
        print("추천수:", recom)
        
        rogo.append({"title":title, "count":count, "recommend":recom})

rogo
    
    
```

4) 추출한 데이터(리스트) 데이터 프레임으로 변환한 후 엑셀파일로 저장

```python
import pandas as pd
```

```python
list_index = ['title','count','recommend']
data = pd.DataFrame(rogo, columns=list_index)
data.to_csv('./rogo_gall.csv', encoding='utf-8')
```

### 2. 같은 코드 반복으로 여러 페이지 크롤링하기. 

1) 1번 검색했을 때 2~3페이지가 검색되고, 각 1번씩의 검색에서 페이지 당 데이터를 추출해야 하므로 규칙을 찾아 반복문 작성하기

```python
#page, search_pos 를 바꿔 여러 페이지를 추출할 수 있다는 규칙 발견

url = "https://gall.dcinside.com/board/lists?id=skwyverns"
#url2 = "https://gall.dcinside.com/board/lists/?id=skwyverns&page=2&search_pos=&s_type=search_subject_memo&s_keyword=%EB%A1%9C%EA%B3%A0"
#"search_pos":"-9765451" 10000씩 줄어듬.

rogo1 = []

#페이지를 이용해 반복문 작성(원래 작성했던 반복문 포함)
for j in range(1,2 + 1):
    params = {"page":j,"search_pos":"-9765451","s_type":"search_subject_memo","s_keyword":encode}
    headers = {"User-Agent":"-"}
    resp1 = requests.get(url,params=params,headers=headers)
    #print(resp1)
    
    parse1 = BeautifulSoup(resp1.content,"html.parser")
    content1 = parse1.find("tbody")
    list_tag1 = content1.find_all("tr")
    
    for i in list_tag1:
        if i.find("em",{"class":"icon_img icon_notice"}):
            pass
        elif i.find("em",{"class":"icon_img icon_survey"}):
            pass
        else:
            print("-" * 10)

            title_tag1 = i.find("a")
            title1 = title_tag1.text
            print("제목:", title1)

            count_tag1 = i.find("td",{"class":"gall_count"})
            count1 = count_tag1.text
            print("조회수:", count1)

            recommend_tag1 = i.find("td",{"class":"gall_recommend"})
            recom1 = recommend_tag1.text
            print("추천수:", recom1)
        
            rogo1.append({"title":title1, "count":count1, "recommend":recom1})

```
